```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
labels = np.load("hiv_label.npy", allow_pickle=True)
dict = {'NK cell': 6, 'DCs': 7, 'T cell': 1, 'B cell': 4, 'CTLs': 5, 'monocyte': 2, 'plasmablast': 3}
color = np.zeros(labels.shape)
print(color.shape)
for i in range(color.shape[0]):
    color[i] = dict[labels[i]]-1
```

```python
plt.figure(figsize=(20,16))
fontsize = 26

for i in range(5):
    plt.subplot(4,5,i+1)
    X = np.load('diff_run_results_server/kazer_t-SNE.npy')[i,:]
    plt.scatter(X[:,0], X[:,1], s=0.01, c=color, cmap="RdYlGn_r")
    plt.title('t-SNE', fontsize=fontsize)
    plt.axis('off')
    
for i in range(5):
    plt.subplot(4,5,i+1 + 5)
    X = np.load('diff_run_results_server/kazer_UMAP.npy')[i,:]
    plt.scatter(X[:,0], X[:,1], s=0.01, c=color, cmap="RdYlGn_r")
    
    if i == 0:
        a = 10.8
        b = 6.5
        x = np.arange(-6 + a, 6 + a, 0.1)
        y = np.arange(-6 + b, 6 + b, 0.1)
        x, y = np.meshgrid(x,y)
        plt.contour(x, y, 2.3*(x-a)**2 + 2.5*(y-b)**2, [4], colors='black', linewidths=1.5)

        a = 5.4
        b = 0.2
        x = np.arange(-6 + a, 6 + a, 0.1)
        y = np.arange(-6 + b, 6 + b, 0.1)
        x, y = np.meshgrid(x,y)
        plt.contour(x, y, 4.3*(x-a)**2 + 4.5*(y-b)**2, [4], colors='black', linewidths=1.5)
    
    if i == 3:
        a = 13.1
        b = 6
        x = np.arange(-6 + a, 6 + a, 0.1)
        y = np.arange(-6 + b, 6 + b, 0.1)
        x, y = np.meshgrid(x,y)
        plt.contour(x, y, 2.3*(x-a)**2 + 2.5*(y-b)**2, [4], colors='black', linewidths=1.5)

        a = 7.2
        b = -8.
        x = np.arange(-6 + a, 6 + a, 0.1)
        y = np.arange(-6 + b, 6 + b, 0.1)
        x, y = np.meshgrid(x,y)
        plt.contour(x, y, 4.3*(x-a)**2 + 4.5*(y-b)**2, [4], colors='black', linewidths=1.5)


    plt.title('UMAP', fontsize=fontsize)
    plt.axis('off')
    
for i in range(5):
    plt.subplot(4,5,i+1 + 10)
    X = np.load('diff_run_results_server/kazer_TriMap.npy')[i,:]
    plt.scatter(X[:,0], X[:,1], s=0.01, c=color, cmap="RdYlGn_r")
    plt.title('TriMap', fontsize=fontsize)
    plt.axis('off')
    
for i in range(5):
    plt.subplot(4,5,i+1 + 15)
    X = np.load('diff_run_results_server/kazer_PaCMAP.npy')[i,:]
    plt.scatter(X[:,0], X[:,1], s=0.01, c=color, cmap="RdYlGn_r")
    plt.title('PaCMAP', fontsize=fontsize)
    plt.axis('off')
    
plt.tight_layout()
plt.savefig('diff_run')
plt.show()
```
