```python
from matplotlib.patches import Ellipse, Circle
import matplotlib.pyplot as plt


pacmap_70 = pd.read_csv('norm_70_pacmap_redo.csv')
pacmap_70 = np.array(pacmap_70)

umap_70 = pd.read_csv('norm_70_umap.csv')
umap_70 = np.array(umap_70)

tsne_70 = pd.read_csv('norm_70_tsne.csv')
tsne_70 = np.array(tsne_70)

#-----------------------------------

pacmap_70_count = pd.read_csv('count_70_pacmap_redo.csv')
pacmap_70_count = np.array(pacmap_70_count)

umap_70_count = pd.read_csv('count_70_umap.csv')
umap_70_count = np.array(umap_70_count)

tsne_70_count = pd.read_csv('count_70_tsne.csv')
tsne_70_count = np.array(tsne_70_count)

#-----------------------------------

gml_pacmap = pd.read_csv('gml_pacmap_csv.csv')
gml_pacmap = np.array(gml_pacmap)

gml_umap = pd.read_csv('gml_umap_csv.csv')
gml_umap = np.array(gml_umap)

gml_tsne = pd.read_csv('gml_tsne_csv.csv')
gml_tsne = np.array(gml_tsne)
```
